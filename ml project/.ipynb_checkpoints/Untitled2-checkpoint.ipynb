## Data visulalization

### Import packages and load dataset

```python
import pandas as pd
import numpy as np                   
import matplotlib.pyplot as plt    
import seaborn as sns              
%matplotlib inline
import datetime
from dateutil.relativedelta import relativedelta
from datetime import date
from datetime import time
```

```python
# Load the dataset into a pandas dataframe
car_dataset = pd.read_csv('car_dataset.csv', sep=',',low_memory=False, encoding ='latin-1')
```

### Data cleaning

### Remove all insignificant columns

```python
car_dataset.info()
```

```python
car_dataset.head(5)
```

```python
car_dataset.drop(['New_Price'],axis=1,inplace=True)
car_dataset.columns
```

### remove all duplicates values

```python
car_dataset
```

### remove blank rows

```python
car_dataset.dropna(axis=0,how='all',inplace=True)
```

```python
car_dataset
```

```python
print(car_dataset.shape)
```

### remove standarad type(NaN) missing value with 0

```python
car_dataset.fillna(0,inplace=True)
```

```python
car_dataset
```

### Change some column names using rename() function

```python
# this step is just to simplify dataset attributes
car_dataset.rename(columns={'Year':'Year_of_Registration', 'Kilometers_Driven':'Odometer_Reading','Transmission':'Transmission_Type',
        'Mileage(kmpl)':'Mileage_in_kmpl','Engine(CC)':'Engine_in_cc','Power(bhp)':'Power_in_bhp','Seats':'Number_of_seats','Price':'Price_in_lakh'},inplace=True)
car_dataset.columns
```

## variables by their type: quantitative vs.categorical


```python
#read dataset
car_dataset_dup=car_dataset[['Fuel_Type','Transmission_Type','Odometer_Reading','Mileage_in_kmpl','Engine_in_cc','Power_in_bhp','Number_of_seats','Price_in_lakh']]
car_dataset_dup.head()
```

```python
car_dataset_dup.info()
```

## Handle categorical variables with numerically coded values

```python
car_dataset.dtypes
```

we can see that :-
1. Fuel_Type and Transmission_Type are Categorical variables
2. Odometer_Reading and Mileage_in_kmpl are Quantitaive variables.
so we should modify it accordingly
handling categorical variables with numerical variables -:
1. For Fuel_Type, create new column "Fuel_Type_id" which handles categorical data as numerical data.
 0 <- CNG
 1 <- Diesel
 2 <- Electric
 3 <- LPG
 4 <- Petrol
2. For Transmission_Type, create new column "Transmission_Type_id" which
handles categorical data as numerical data.
 0 <- Automatic
 1 <- Manual


```python
car_dataset_dup['Fuel_Type_id']=pd.Categorical(car_dataset_dup.Fuel_Type).codes
car_dataset_dup['Transmission_Type_id']=pd.Categorical(car_dataset_dup.Transmission_Type).codes
car_dataset_dup.head(10)
```

## Univariate Analysis


```python
Rang=['lightskyblue','lightcoral']
plt.pie(car_dataset_dup.groupby('Transmission_Type').count()['Mileage_in_kmpl'],
        colors=Rang,
        autopct='%1.2f%%',
        pctdistance=0.5,
        shadow=True,
        wedgeprops={"edgecolor":"0",'linewidth': 1,'linestyle': 'solid', 'antialiased': True}
       )
plt.axis('equal') 
Transmission=['Automatic','Manual']
plt.legend(Transmission,title="Transmission",loc='upper right')
plt.title('Per% of data with Transmission',color='black',size=15)
plt.show()
car_dataset_dup['Fuel_Type'].value_counts().plot.bar(title='Bar Graph for Fuel_Type')
plt.grid(color='black')
plt.show()
```

```python
col_names = ['Mileage_in_kmpl','Odometer_Reading']
temp, x = plt.subplots(len(col_names), figsize=(16,15))
for i,col_val in enumerate(col_names):
    sns.distplot(car_dataset_dup[col_val], hist=True, ax=x[i])
    x[i].set_title('Freq dist '+col_val, fontsize=20)
    x[i].set_xlabel(col_val, fontsize=16)
    x[i].set_ylabel('Count', fontsize=16)
plt.show()
```
